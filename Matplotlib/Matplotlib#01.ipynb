```python
import matplotlib.mlab as mlab
mu = 100
sigma = 20
x = mu + sigma*np.random.randn(10000)
num_bins = 50 ## Áü©ÂΩ¢Êù°ÁöÑ‰∏™Êï∞
# normed=1Ë°®Á§∫ÁîªÁöÑÊòØÊ¶ÇÁéáÂØÜÂ∫¶ÔºåÂíå‰∏∫1ÔºõcolorÔºöÁü©ÂΩ¢Êù°È¢úËâ≤ÔºõalphaÔºöËâ≤Ê∑±ÂèÇÊï∞
n, bins, patches = plt.hist(x, num_bins, normed=1, color='green', alpha=0.6)
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('histogram')
plt.show()
```

```python
## 3dÁ©∫Èó¥Êõ≤Èù¢ax.plot_surface()
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5.5, 0.1) # XÂùêÊ†á 
Y = np.arange(-5.5, 0.1) # YÂùêÊ†á
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) ## ËÆ°ÁÆóÂá∫ZÂùêÊ†á
surf = ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm,linewidth=0,antialiased=False)
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10)) #ËÆæÁΩÆZËΩ¥Ê†áÂ∫¶
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
